```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#You can call and use the .head functions to test the individual variables, for the sake of cleanliness, removed the code for this
schoolTotal=school_data["school_name"].count()

studentTotal = student_data["student_name"].count()


budgetTotal = school_data["budget"].sum()


mathAvg=student_data[("math_score")].mean()


readingAvg = student_data[("reading_score")].mean()
readingAvg

schoolAvgScore = ((mathAvg +readingAvg)/2)


mathPassPerc = (school_data_complete["math_score"]>70).sum() / school_data_complete["math_score"].count() *100


readingPassPerc = (school_data_complete["reading_score"]>70).sum() / school_data_complete["reading_score"].count() *100

#Turning out the results in first iteration
districtSummary = pd.DataFrame({"Total Schools":[schoolTotal],"Total Students":[studentTotal],"Total Budget":[budgetTotal],"Average Math Score":[mathAvg],"Average Reading Score":[readingAvg],"% Passing Math":[mathPassPerc],
                                "% Passing Reading":[readingPassPerc],"% Overall Pass Rate":[schoolAvgScore]})

#then transferred it over here to help map and clean/format the data
districtSummaryDF = pd.DataFrame(districtSummary, columns=["Total Schools","Total Students","Total Budget","Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Pass Rate"])


districtSummaryDF["Total Students"]=districtSummaryDF["Total Students"].map("{:,}".format)
districtSummaryDF["Total Budget"]=districtSummaryDF["Total Budget"].map("{:,}".format)
districtSummaryDF["Average Math Score"]=districtSummaryDF["Average Math Score"].map("{:,.2f}".format)

districtSummaryDF.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Setting index to be tied to the name and its type, and then pulling other info about the schools such as 
schoolType = school_data.set_index(["school_name"])["type"]
studentsPerSchool = school_data_complete["school_name"].value_counts()
schoolBudget = school_data_complete.groupby(["school_name"])["budget"].mean()
perstudentBudget =schoolBudget/studentsPerSchool

#score average of math and reading by school plus overall passing rate of the two subjects
mathAvgPerSchool=school_data_complete.groupby(["school_name"])["reading_score"].mean()
passMathScore=school_data_complete.loc[school_data_complete["math_score"]>=70]
mathScoreGroup = passMathScore["school_name"].value_counts()
percentMathPass=mathScoreGroup/studentsPerSchool*100

readingAvgPerSchool=school_data_complete.groupby(["school_name"])["reading_score"].mean()
passReadingScore = school_data_complete.loc[school_data_complete["reading_score"]>=70]
readingScoreGroup = passReadingScore["school_name"].value_counts()
percentReadingPass =readingScoreGroup/studentsPerSchool*100

PassingOverallScore = percentMathPass+percentReadingPass/studentsPerSchool
#adds the new variables we made
schoolSummaryDF = pd.DataFrame({"School type":schoolType,"Total Students":studentsPerSchool,"Total School budget":schoolBudget,"Per Student Budget":perstudentBudget,"Average Math Score":mathAvgPerSchool,
                                "Average Reading Score":readingAvgPerSchool,"% Passing Math":percentMathPass,"% Passing Reading":percentReadingPass,"% Overall Passing Rate":PassingOverallScore})
schoolSummaryDF.head(15)
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#using the sortvalue function to show the top 5 results
topSchoolDF = schoolSummaryDF.sort_values(["% Passing Math","% Passing Reading","% Overall Passing Rate"],ascending=False)
topSchoolDF.head()
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
#using the  sortvalue function to call in the worst schools 
bottomSchoolDF = schoolSummaryDF.sort_values("% Overall Passing Rate")
bottomSchoolDF.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#grabbing the grade with loc and grouping those students by their school name and math score, then becomes its own dataframe with tailored results
grade9MathScores=student_data.loc[student_data["grade"]=="9th"].groupby(["school_name"])["math_score"].mean()
grade10MathScores=student_data.loc[student_data["grade"]=="10th"].groupby(["school_name"])["math_score"].mean()
grade11MathScores=student_data.loc[student_data["grade"]=="11th"].groupby(["school_name"])["math_score"].mean()
grade12MathScores=student_data.loc[student_data["grade"]=="12th"].groupby(["school_name"])["math_score"].mean()

mathByGradeDF = pd.DataFrame({"9th":grade9MathScores,"10th":grade10MathScores, "11th":grade11MathScores,"12th":grade12MathScores})
mathByGradeDF
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#grabbing the grade with loc and grouping those students by their school name and reading score, then becomes its own dataframe with tailored results
grade9ReadingScores=student_data.loc[student_data["grade"]=="9th"].groupby(["school_name"])["reading_score"].mean()
grade10ReadingScores=student_data.loc[student_data["grade"]=="10th"].groupby(["school_name"])["reading_score"].mean()
grade11ReadingScores=student_data.loc[student_data["grade"]=="11th"].groupby(["school_name"])["reading_score"].mean()
grade12ReadingScores=student_data.loc[student_data["grade"]=="12th"].groupby(["school_name"])["reading_score"].mean()

readingByGradeDF = pd.DataFrame({"9th":grade9ReadingScores,"10th":grade10ReadingScores, "11th":grade11ReadingScores,"12th":grade12ReadingScores})
readingByGradeDF
```
